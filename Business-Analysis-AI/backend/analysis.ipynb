```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("shopping_trends.csv")
```

```python
item_sales = df["Item Purchased"].value_counts()

print("Top Selling Items:\n", item_sales)
```

```python
customer_prefs = df.groupby(["Gender", "Age"])["Item Purchased"].value_counts()
print("Customer Preferences:\n", customer_prefs)
```

```python
category_spending = df.groupby("Category")["Purchase Amount (USD)"].mean().sort_values(ascending=False)
print("Avg Spending by Category:\n", category_spending)
```

```python
regional_orders = df["Location"].value_counts()
print("Orders by Region:\n", regional_orders)
```

```python
# *2. Customer Preferences (Bar Chart by Gender)*
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Item Purchased", hue="Gender", palette="muted", order=df["Item Purchased"].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel("Items")
plt.ylabel("Number of Purchases")
plt.title("Customer Preferences by Gender")
plt.legend(title="Gender")
plt.show()
```
